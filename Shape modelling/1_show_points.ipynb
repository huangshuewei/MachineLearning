```python
"""
Example of loading points from a matrix in a text file.
Each row is a set of points in format: x0 y0 x1 y1 ....

"""

import matplotlib.pyplot as plt
import numpy as np

# Plot the points (assumed stored as single row {x0,y0,x1,y1...})
def plot_points(plt, points, style):
    pts=points.reshape(-1,2)
    plt.plot(pts[:,0],pts[:,1],style)
    
shapes=np.genfromtxt("pelvis_shapes(2).txt")

n_shapes=int(shapes.shape[0])
print("Number of shapes", n_shapes)
n_pts=int(shapes.shape[1]/2)
print("Number of points", n_pts)

plot_points(plt,shapes[0,:],"o")
plt.title("First shape")
plt.show()

# Plot all the points to show the scatter
for i in range(n_shapes):
    plot_points(plt,shapes[i,:],"+")
    
plt.title("Scatter of all points")
plt.show()
```

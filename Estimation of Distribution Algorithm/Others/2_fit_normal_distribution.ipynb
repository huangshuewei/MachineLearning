```python
"""
Estimate parameters of a normal distribution.
"""
import numpy as np
import matplotlib.pyplot as plt

mean=np.array([1,2])

# Generate a random positive definite matrix
A=np.random.randn(2,2)
cov=A.T@A

print("Covariance:\n",cov)

n_samples=1000
x=np.random.multivariate_normal(mean,cov,n_samples)

plt.plot(x[:,0],x[:,1],"x",color="green")  
# plt.axis('equal')
plt.show()

# Estimate parameters
est_mean=x.mean(0)
print("Estimate of mean: ",est_mean)

# Variables are in columns, so rowvar=False
est_cov=np.cov(x,rowvar=False)
print("Estimate of cov:\n",est_cov)

evals,EVecs=np.linalg.eig(est_cov)
print("Eigenvalues of covar: ",evals)
```

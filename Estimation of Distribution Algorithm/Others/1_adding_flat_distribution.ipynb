```python
"""
Sampling from distributions

Code to demonstrate how the sum of samples from flat
distributions leads to a Normal distribution

"""

import numpy as np
import matplotlib.pyplot as plt

print("Generate n samples from adding n_dice 6-digits dice.")
```

```python
n=100

n_dice=1

# n rolls of n_dice dice
r=np.random.randint(1,7, [n,n_dice])

# Add up the rolls
x=np.sum(r,1)
bins=np.arange(n_dice,n_dice*6+2,1)-0.5
plt.hist(x,bins)
plt.title("one dice and do 100 times ")
plt.show()
```

```python
# Task 1: Observe how theoretically flat distribution isn't until n is large
# Se large n
n=100000

# n rolls of n_dice dice
r=np.random.randint(1,7, [n,n_dice])

# Add up the rolls
x=np.sum(r,1)
bins=np.arange(n_dice,n_dice*6+2,1)-0.5
plt.hist(x,bins)
plt.title("one dice and do 100000 times ")
plt.show()
```

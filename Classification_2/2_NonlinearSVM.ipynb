```python
"""
Simple example of non-linear Support Vector Machine
"""

from sklearn import svm
from sklearn.datasets.samples_generator import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# Display contour plot of f at all points (x[i],y[j])
def contour_plot1(x,y,f,n_levels):
    xx, yy = np.meshgrid(x, y)
    p=np.hstack([xx.reshape(-1,1),yy.reshape(-1,1)])
    fp=f(p)
    plt.contourf(x,y,fp.reshape(xx.shape),n_levels,cmap="gray")
```

```python
# Generate samples from two classes
X, y = make_blobs(n_samples=100, centers = [(2, 1),(2,8), (3.5, 4)], 
                  n_features=2,random_state=17)
# Set blobs 0,1 to class 0, blob 2 to class 1
y[y<=1]=0
y[y>1]=1

print("kernel='rbf', gamma=0.5, C=1")
clf = svm.SVC(kernel='rbf', gamma=0.5, C=1)
#clf = svm.SVC(kernel='linear')
#clf = svm.SVC(kernel='poly',degree=2,coef0=1.0)
clf.fit(X, y)

print("Number of support vectors: ",clf.support_vectors_.shape[0])

# Create a grid of points
px = np.arange(X[:,0].min()-.5, X[:,0].max()+.5, 0.1)
py = np.arange(X[:,1].min()-.5, X[:,1].max()+.5, 0.1)


# Use different colours for different classes
X0=X[y==0] # Select samples for class 0
X1=X[y==1] # Select samples for class 1
plt.plot(X0[:,0],X0[:,1],"+",color="orange")
plt.plot(X1[:,0],X1[:,1],"o",color="green")
contour_plot1(px,py,clf.predict,15)

# Highlight the support vectors with coloured circles
sp=clf.support_vectors_
plt.scatter(sp[:,0],sp[:,1],s=90, facecolors='none', edgecolors="red")
plt.title("Estimated class at each point")
plt.show()
```

```python
# Try with a linear kernel
print("kernel='linear'")
clf = svm.SVC(kernel='linear')
# clf = svm.SVC(kernel='poly',degree=2,coef0=1.0)
clf.fit(X, y)

print("Number of support vectors: ",clf.support_vectors_.shape[0])

# Create a grid of points
px = np.arange(X[:,0].min()-.5, X[:,0].max()+.5, 0.1)
py = np.arange(X[:,1].min()-.5, X[:,1].max()+.5, 0.1)


# Use different colours for different classes
X0=X[y==0] # Select samples for class 0
X1=X[y==1] # Select samples for class 1
plt.plot(X0[:,0],X0[:,1],"+",color="orange")
plt.plot(X1[:,0],X1[:,1],"o",color="green")
contour_plot1(px,py,clf.predict,15)
# Highlight the support vectors with coloured circles
sp=clf.support_vectors_
plt.scatter(sp[:,0],sp[:,1],s=90, facecolors='none', edgecolors="red")
plt.title("Estimated class at each point")
plt.show()
```
